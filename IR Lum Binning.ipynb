```python
def z_splitter(bins, data):
    bans = {}
    if type(bins) == int:
        z_max = max([row[0] for row in data])
        interval = z_max/bins
        bons = []
        for i in range(bins):
            bons.append((i+1)*interval)
            bans[(i+1)*interval] = []
        bins = bons
    else:
        for b in bins:
            bans[b] = []
    for row in data:
        for b in bins:
            if row[0] < b:
                bans[b].append(row)
                break
    return bans
```
